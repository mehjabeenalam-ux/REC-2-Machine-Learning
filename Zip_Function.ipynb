Basic Use Cases of Zip Functions

Zip Function (takes 2 or more sequences and converts into a Tuple)

Example 1 with Similar Data Types

```python
shirts = ["Red", "Blue", "Green"]
pants  = ["Black", "Grey", "White"]

outfits = list(zip(shirts, pants))
print(outfits)
```

Example 2 with Multiple Data Types

```python
names = ["Alice", "Bob", "Charlie"]
ages  = [25, 30, 35]

zipped = list(zip(names, ages))
print(zipped)
```

Example 3 Unpacking with * (very common pattern) through running For Loops

```python
for name, age in zip(names, ages):
    print(f"{name} is {age} years old")
```

Example 4 Working on more than 2 iterables with different data types

```python
letters = ['a', 'b', 'c']
numbers = [1, 2, 3]
symbols = ['!', '@', '#']

print(list(zip(letters, numbers, symbols)))
```

Example 5 Unequal lengths â†’ stops at shortest

```python
x = [1, 2, 3, 4]
y = [10, 20]
print(list(zip(x, y)))
```

Processing Sales Data with Zip Functions

```python
products = ["Laptop", "Phone", "Tablet"]
prices   = [1200, 800, 450]
qty_sold = [15, 30, 25]

# Calculate total revenue per product in one clean loop
for product, price, quantity in zip(products, prices, qty_sold):
    revenue = price * quantity
    print(f"{product}: ${revenue:,} revenue")
```

Zip Function with User Inputs

```python
# Ask user for first list (e.g. names of products)
print("Enter some names (separate with commas):")
names_input = input().strip()
names = [name.strip() for name in names_input.split(',')]

# Ask for second list (e.g. ages)
print("Enter their sales quantity (same order, separate with commas):")
sales_input = input().strip()
sales = [int(sale.strip()) for sale in sales_input.split(',')]

# Now zip them together!
print("\nHere are your paired results:")
for name, sale in zip(names, sales):
    print(f"Total Quantity Sold for {name} is {sale} tons")

# # Bonus: Show as a nice list too
paired = list(zip(names, sales))
print("\nAll pairs:", paired)
```
