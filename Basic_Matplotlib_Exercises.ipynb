Basic Plotting and X, Y Coordinates Placing

```python
from matplotlib import pyplot as plt

#Data
X = [1, 2, 3, 4, 5]
Y = [0.90, 0.70, 0.55, 0.45, 0.40]

#Plot
plt.plot(X, Y) # Plot the data
```

Decoration

```python
plt.figure(figsize=(15, 4)) # Set the figure size
plt.plot(X, Y) # Plot the data
```

Adding Title to Plots

```python
plt.figure(figsize=(15, 4)) # Set the figure size
plt.title("X vs Y Plot") # Add a title
plt.plot(X, Y) # Plot the data
```

Labelling X and Y axis

```python
plt.figure(figsize=(15, 4)) # Set the figure size
plt.title("X vs Y Plot") # Add a title
plt.xlabel("X values") # Label for X-axis
plt.ylabel("Y values") # Label for Y-axis
plt.plot(X, Y) # Plot the data
```

Adding Grid details

```python
plt.figure(figsize=(15, 4)) # Set the figure size
plt.title("X vs Y Plot") # Add a title
plt.xlabel("X values") # Label for X-axis
plt.ylabel("Y values") # Label for Y-axis
plt.plot(X, Y) # Plot the data
plt.grid(True, alpha=0.3)
```

Multiple Plots in Single Diagram

```python
x = [1, 2, 3, 4, 5]

y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 4, 8, 16]
y3 = [10, 9, 7, 6, 5]

plt.plot(x, y1, label="Line 1") # Plot line 1
plt.plot(x, y2, label="Line 2") # Plot line 2
plt.plot(x, y3, label="Line 3") # Plot line 3

plt.title("Multiple Line Graph Without Legend") # Add title
plt.xlabel("x") # Add labels
plt.ylabel("y") # Add labels
plt.grid(True, alpha=0.3)   # Add grid
plt.show()  # Display the plot
```

Multiple Line Graph With Legend

```python
plt.plot(x, y1, label="Line 1") # Plot line 1
plt.plot(x, y2, label="Line 2") # Plot line 2
plt.plot(x, y3, label="Line 3") # Plot line 3

plt.title("Multiple Line Graph With Legend") # Add title
plt.xlabel("x") # Add labels
plt.ylabel("y") # Add labels
plt.legend()    # Show legend
plt.grid(True, alpha=0.3)   # Add grid
plt.show()  # Display the plot
```

Histogram

```python
data = [1.2, 2.3, 2.9, 3.1, 3.8, 4.0, 4.2, 5.5, 6.1, 6.8]

plt.figure(figsize=(7, 4))  # Set the figure size
plt.hist(data, bins=6, edgecolor="black")   # Create histogram
plt.title("Histogram")  # Add title
plt.xlabel("Value") # Label for X-axis
plt.ylabel("Count") # Label for Y-axis
plt.grid(True, alpha=0.3)   # Add grid
plt.show()  # Display the plot
```

Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [2.1, 2.9, 3.7, 4.2, 5.1, 5.9]

plt.figure(figsize=(7, 4))
plt.scatter(x, y)
plt.title("Scatter Plot", )
plt.xlabel("Feature (x)")
plt.ylabel("Target (y)")
plt.grid(True, alpha=0.3)
plt.show()
```

Box Plot

```python
group_a = [10, 12, 13, 12, 11, 14, 30]   # 30 is an outlier
group_b = [8, 9, 10, 11, 9, 10, 12]

plt.figure(figsize=(7, 4))
plt.boxplot([group_a, group_b], tick_labels=["Group A", "Group B"])
plt.title("Box Plot")
plt.ylabel("Value")
plt.grid(True, axis="y", alpha=0.3)
plt.show()
```

Heatmap

```python
matrix = [
    [1.0, 0.8, 0.2],
    [0.8, 1.0, 0.5],
    [0.2, 0.5, 1.0]
]

plt.figure(figsize=(6, 5))
plt.imshow(matrix)
plt.title("Heatmap")
plt.xticks([0, 1, 2], ["F1", "F2", "F3"])
plt.yticks([0, 1, 2], ["F1", "F2", "F3"])
plt.colorbar(label="Value")
```

Bar Chart

```python
classes = ["Cat", "Dog", "Bird"]
counts = [50, 35, 15]

plt.figure(figsize=(7, 4))
plt.bar(classes, counts, edgecolor="black")
plt.title("Bar Chart (Class Counts)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.grid(True, axis="y", alpha=0.3)
plt.show()
```

Image

```python
import matplotlib.image as mpimg

img = mpimg.imread("minions.jpg") # path to your image file
plt.figure(figsize=(8, 5)) # Set the figure size
plt.imshow(img) # Display the image
plt.axis("off")          # hide axes
plt.title("Image")
plt.show()
```

Multiple Images Side by Side

```python
img1 = mpimg.imread("minions.jpg")
img2 = mpimg.imread("minions.jpg")

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].imshow(img1); ax[0].axis("off"); ax[0].set_title("Image 1")
ax[1].imshow(img2); ax[1].axis("off"); ax[1].set_title("Image 2")
plt.show()
```

Multiple Images (Grid)

```python
# load images
img1 = mpimg.imread("minions.jpg")
img2 = mpimg.imread("minions.jpg")
img3 = mpimg.imread("minions.jpg")
img4 = mpimg.imread("minions.jpg")

# create 2x2 grid
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

ax[0, 0].imshow(img1); ax[0, 0].set_title("Image 1 [0,0]"); ax[0, 0].axis("off")
ax[0, 1].imshow(img2); ax[0, 1].set_title("Image 2 [0,1]"); ax[0, 1].axis("off")
ax[1, 0].imshow(img3); ax[1, 0].set_title("Image 3 [1,0]"); ax[1, 0].axis("off")
ax[1, 1].imshow(img4); ax[1, 1].set_title("Image 4 [1,1]"); ax[1, 1].axis("off")

plt.tight_layout()
plt.show()
```
