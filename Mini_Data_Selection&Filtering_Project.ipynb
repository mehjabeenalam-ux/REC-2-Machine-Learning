Data Selection and Filtering Mini Example from ChatGPT

```python
import pandas as pd
```

Sample employee dataset

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 40],
    'salary': [70000, 120000, 85000, 95000],
    'department': ['IT', 'Sales', 'IT', 'Sales']
})

print(df)
```

1. Boolean filtering – most common & fastest

```python
high_salary = df[df['salary'] > 90000]                  # Employees earning > $90k
print("High earners:\n", high_salary)

```

2. .loc[] with condition – label/boolean style (very flexible)

```python
senior_staff = df.loc[df['age'] > 30, ['name', 'salary']]  # Only name & salary for age > 30
print("\nSenior staff:\n", senior_staff)
```

3. .query() – clean SQL-like syntax (great for complex conditions)

```python
it_high_earners = df.query("department == 'IT' and salary > 80000")
print("\nIT high earners:\n", it_high_earners)
```

4. .iloc[] – position-based slicing (integer index)

```python
middle_rows = df.iloc[1:3]                             # Rows 1 to 2 (0-based, exclusive end)
print("\nMiddle rows:\n", middle_rows)
```

5. .isin() – membership check (filter by list of values)

```python
sales_or_it = df[df['department'].isin(['Sales', 'IT'])]
print("\nSales or IT department:\n", sales_or_it)
```
