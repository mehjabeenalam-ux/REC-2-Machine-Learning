How to Calculate Z-Score in Pandas (Practical Code)

Import Pandas Library

```python
import pandas as pd
```

Sample Employee Dataset

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'salary': [70000, 120000, 85000, 95000, 45000]
})

print(f"Employee Dataset {df}")
```

Shape and Size of the Employee Dataset

```python
print(df.shape)
print(df.size) #Row*Column
```

Summary Statistics of the Employee Dataset

```python
df.describe().round((2))
```

Calculate Z-score for salary column

```python
df['z_salary'] = (df['salary'] - df['salary'].mean()) / df['salary'].std()

print(df[['name', 'salary', 'z_salary']].round(2))
```

Sorting the Z Score of the Employee Dataset

Sort Z-scores from lowest to highest (smallest → largest)

```python
sorted_low_to_high = df.sort_values('z_salary')
print("Sorted Z-scores (lowest to highest):")
print(sorted_low_to_high[['name', 'salary', 'z_salary']].round(3))
```

Interpretation of the above Z Score

- Bob has z_salary = 1.32 → salary is 1.32 standard deviations above the mean (high earner, possible outlier on the high side)

- Eve has z_salary = -1.36 → salary is 1.36 standard deviations below the mean (low earner)

- Charlie is very close to 0 → salary is near the average

Rule of thumb:

- |Z| > 3 → strong outlier; |Z| > 2 → moderate outlier (common thresholds)

Lowest Salary Outlier

```python
outlier_salary1 = round(df['z_salary'].min(), 2)
print("Lowest Salary Outlier to the mean (Z-score nearest 0):\n", outlier_salary1)
```

Highest Salary Outlier

```python
outlier_salary = round(df['z_salary'].max(), 2)
print("Highest Salary Outlier to the mean (Z-score nearest 0):\n", outlier_salary)
```

Row with Z-score closest to the mean

```python
closest_to_mean = df.loc[df['z_salary'].abs() == df['z_salary'].abs().min()]
print("Row closest to the mean (Z-score nearest 0):\n", closest_to_mean)
```
