List Comprehension

```python
celsius = [0, 10, 20, 30]
fahrenheit = [(c * 9/5) + 32 for c in celsius]
print(fahrenheit)
```

Mapping Function

```python
temp_list_celcius = [10, 20, 30, 40]

fahrenheit = list(map(lambda c: c * 9/5 + 32, temp_list_celcius))

print(fahrenheit)
```

Supersonic NumPy Arrays

```python
import numpy as np

X_Celcius = np.array([10, 20, 30, 40]) #List of temperatures in Celcius

CelciusToFahrenheit_Calculator = (X_Celcius * (9/5)) + 32

print(CelciusToFahrenheit_Calculator)
```
