Profit and Loss Statements using Zip Functions

Create a list of items, cost price, sale prices, quantities in stock and quantities sold

```python
items_grocery = ["Rice", "Sugar", "Salt", "Eggs", "Oil", "Flour"]
sales_prices = [2, 3, 4, 5, 6, 7]
quantities_sold_tons = [100, 200, 300, 400, 500, 600]

print(items_grocery)
print(sales_prices)
print(quantities_sold_tons)

print(f"Grocery Items Sold: {items_grocery}")
print(f"Grocery Sales Prices in USD of: {items_grocery} are {sales_prices}")
print(f"Grocery Quantities sold in USD : {items_grocery} at {sales_prices} are {quantities_sold_tons} tons")

```

Let's Zip them into one iterable

```python
all_items = list(zip(items_grocery, sales_prices, quantities_sold_tons))
print(all_items)
```

Mini Sales Report through a For Loop

```python
for item, sale, quantity in (zip(items_grocery, sales_prices, quantities_sold_tons)):
  print(item, sale, quantity)
```

For Loop with Formatted Strings for Mini Sales Report

```python
for item, sale, quantity in (zip(items_grocery, sales_prices, quantities_sold_tons)):
  print(f"{item} Sales Price USD {sale} per Kilogram, Monthly Sales {quantity} tons")
```

Revenue Calculations done 2 ways


```python
Revenue = [sale*quantity for sale, quantity in zip(sales_prices, quantities_sold_tons)]

print(Revenue)

Total_Revenue = []

for sale, quantity in zip(sales_prices, quantities_sold_tons):
    Total_Revenue.append(sale * quantity)

print(Total_Revenue)
```

Sum of Revenue

```python
Sum_Revenue_All_Items = sum(Total_Revenue)
print(Sum_Revenue_All_Items)
```

Expenses (Enter Expenses for Rent, Utlity Bills, Salaries, Others)

```python
Expenses = [2000, 3000, 1000, 2000]
print(Expenses)
```

Adding up Expenses

```python
Total_Expenses = sum(Expenses)
print(Total_Expenses)
```

Profit/Loss

```python
Profit = Sum_Revenue_All_Items - Total_Expenses
print(Profit)

```

Ternary Operator for Auto calculating Profit/Loss

```python
Net_result = "Net Profit" if Profit > 0 else "Net Loss"
print(Net_result)
```
