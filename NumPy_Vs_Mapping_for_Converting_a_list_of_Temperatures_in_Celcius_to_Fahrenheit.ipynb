Arrays for Temperature Converter

```python
import numpy as np

X_Celcius = np.array([10, 20, 30, 40]) #List of temperatures in Celcius

CelciusToFahrenheit_Calculator = (X_Celcius * (9/5)) + 32

print(CelciusToFahrenheit_Calculator)

```

Mapping Function to convert temperature lists in Celcius to Fahrenheit

```python
temp_list_celcius = [10, 20, 30, 40]

fahrenheit = list(map(lambda c: c * 9/5 + 32, temp_list_celcius))

print(fahrenheit)
```

NumPy Array Vs Mapping Function Computation for this tiny array of 4 elements

```python
import numpy as np
import timeit

small_list = [10, 20, 30, 40]
small_np = np.array(small_list)

# NumPy version
np_time_small = timeit.timeit(lambda: small_np * (9/5) + 32, number=100000)

# Map version
map_time_small = timeit.timeit(lambda: list(map(lambda c: c * 9/5 + 32, small_list)), number=100000)

print(f"Small data (100,000 runs):")
print(f"NumPy: {np_time_small:.6f} sec")
print(f"Map:   {map_time_small:.6f} sec")

# Larger example: 1 million elements
large_list = list(range(1_000_000))
large_np = np.arange(1_000_000)

np_time_large = timeit.timeit(lambda: large_np * (9/5) + 32, number=100)

map_time_large = timeit.timeit(lambda: list(map(lambda c: c * 9/5 + 32, large_list)), number=100)

print(f"\nLarge data (1M elements, 100 runs):")
print(f"NumPy: {np_time_large:.4f} sec")

print(f"Map:   {map_time_large:.4f} sec")
print(f"NumPy is roughly {(map_time_large / np_time_large):.1f}x faster")
```
