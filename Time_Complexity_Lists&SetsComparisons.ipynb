Example 1: Using List DSA for a Movie Recommendation

```python
all_movies = [
    "Harry Potter", "Avengers", "Inception", "Interstellar",
    "Jumanji", "Frozen", "Avatar", "Titanic"
]

watched_movies = [
    "Harry Potter", "Avatar", "Titanic"
]

recommendations = []

for movie in all_movies:
    if movie not in watched_movies:
        recommendations.append(movie)

print(recommendations)
```

⏳ Why slow?

For every movie → search entire watched list

⏳ Time Complexity: O(n × m)
→ Very slow for Netflix-scale data

Example 2: Using Set DSA for Movie Recommendations

```python
all_movies = {
    "Harry Potter", "Avengers", "Inception", "Interstellar",
    "Jumanji", "Frozen", "Avatar", "Titanic"
}

watched_movies = {
    "Harry Potter", "Avatar", "Titanic"
}

recommendations = all_movies - watched_movies

print(recommendations)

#The exact same code just enclosed inside a curly bracket instead of a third bracket!
```

⚡ Why fast?

Set subtraction is optimized

No loops needed

⚡ Time Complexity: O(n)

Speed Demo: How do we know that set is faster than a list in the movie recommendation codes we wrote above?

```python
import time

movie_list = ["Harry Potter", "Avengers", "Inception", "Interstellar",
    "Jumanji", "Frozen", "Avatar", "Titanic"]

# movie_set = {"Harry Potter", "Avengers", "Inception", "Interstellar",
#     "Jumanji", "Frozen", "Avatar", "Titanic"}

movie_set = set(movie_list)

start_1 = time.time()
"Titanic" in movie_list
print("List time:", time.time() - start_1)

start_2 = time.time()
"Titanic" in movie_set
print("Set time:", time.time() - start_2)

```
