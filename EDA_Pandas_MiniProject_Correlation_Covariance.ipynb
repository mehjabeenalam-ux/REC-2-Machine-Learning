In Pandas, df.isna() and df.isnull() do exactly the same thing — they are identical functions that both return a DataFrame of the same shape as df with True where values are missing (NaN/None) and False otherwise.

Short answer (one sentence):
- There is no difference — isna() and isnull() are aliases (two different names for the same function) in Pandas, so df.isna().sum() and df.isnull().sum() always give identical results.

Step 6: Value Counts and Distribution

 Value Counts and Distribution are among the most important and frequently used tools in the early stages of Exploratory Data Analysis (EDA) when working with pandas.

What do we mean by "distribution" in this context?

When we talk about the distribution of a categorical or discrete variable in EDA, we usually mean:

Which values exist (the unique categories)

How often each value appears (frequency)

How uneven or balanced that frequency is

What proportion each category represents of the total

value_counts() gives you exactly this information — very quickly.

Value_counts() is a pandas Series method that:

- Counts how many times each unique value appears in a column

- Returns the result sorted from most frequent to least frequent by default

Computing Value Counts of the Value Column

```python
print(df["Value"].value_counts()) #Counts occurences of each unique value
print(df["Value"].value_counts(normalize=True)) #Normalized value Counts
```

Computing Value Counts of the Sales Column

```python
print(df["Sales"].value_counts()) #Counts occurences of each unique value
print(df["Sales"].value_counts(normalize=True)) #Normalized value Counts
```

Step 7: Measuring Correlation and Covariance

Sorting Numerical Dataframes

```python
numerical_dataframes = df[['Value', 'Sales']]
print(numerical_dataframes)
```

Correlation Matrix of Numerical Columns *Pearson by Default

```python
print(numerical_dataframes.corr())
```

Correlation *Kendall Method

```python
print(numerical_dataframes.corr(method="kendall"))
```

Correlation *Spearman Method

```python
# df.corr(method="spearman")
print(numerical_dataframes.corr(method="spearman"))
```

 Covariance Matrix of Numerical Columns

```python
print(numerical_dataframes.cov())
```

Correlation (Pearson, Kendall and Spearman) and Covariance together!

```python
print(numerical_dataframes.corr())
print(numerical_dataframes.corr(method="kendall"))
print(numerical_dataframes.corr(method="spearman"))
print(numerical_dataframes.cov())
```
