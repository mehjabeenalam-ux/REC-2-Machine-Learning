Data Overview

```python
df.head(n):Displays first n rows (default=5)
df.tail(n):Displays last n rows (default=5)
df.shape: Returns rows & columns
df.info():Returns columns data types & memory usage

df.columns(): List all columns names
df.index(): Shows Index Range
df.dtypes(): Shows Data Types of all columns

df.describe(): Summary Statistics of numerical columns
df.describe(include=all): Summary Statistics of all columns
```

Checking Missing Values

```python
df.isnull().sum(): Count Missing Values in each column
df.isna().sum(): Count Missing Values in each column

df[df.isnull().any(axis=1)]: Display Rows with Missing Values
df.dropna(): Removes rows with Missing Values

df.fillna(value): Fills Missing Values with a Value

df.fillna(df.median()): Fills Missing Values with a Median
df.fillna(df.mean()): Fills Missing Values with a Mean
df.fillna(df.mode()): Fills Missing Values with a Mode

df.fillna(df.min()): Fills Missing Values with a Minimum Value
df.fillna(df.max()): Fills Missing Values with a Maximum Value

df.fillna(df.std()): Fills Missing Values with a Standard Deviation
df.fillna(df.var()): Fills Missing Values with a Variance

df.fillna(df.corr()): Fills Missing Values with a Correlation
df.fillna(df.cov()): Fills Missing Values with a Covariance

df.fillna(df.skew()): Fills Missing Values with a Skewness
df.fillna(df.kurt()): Fills Missing Values with a Kurtosis


df.interpolate(): Performs Linear Interpolation to fill NAN
```

Checking Duplicate Values

```python
df.duplicated(): Checks Duplicate Values & returns a Boolean Series of Duplicate Values
df.drop_duplicates(): Removes Duplicate Values
df[df.duplicated()]: Displays Duplicate Rows
df.duplicated().sum(): Counts Duplicate Rows
```
