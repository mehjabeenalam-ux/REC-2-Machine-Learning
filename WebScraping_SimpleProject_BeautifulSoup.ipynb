Web Scraping Task: First 5 Books

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Get the webpage
url = "http://books.toscrape.com"
response = requests.get(url)  # Like opening the site
print("Website loaded! Status:", response.status_code)  # 200 means good!

# Step 2: Make "soup" to search easily
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Find all books (they are in <article class="product_pod">)
books = soup.find_all('article', class_='product_pod')

# Step 4: Loop through first 5 books and get title + price
print("\nFirst 5 Books:")
for book in books[:5]:
    title = book.h3.a['title']  # Get book title
    price = book.find('p', class_='price_color').text  # Get price
    print(f"Title: {title}")
    print(f"Price: {price}")
    print("---")
```

Web Scraping Project: First 10 Books

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Get the webpage
url = "http://books.toscrape.com"
response = requests.get(url)  # Like opening the site
print("Website loaded! Status:", response.status_code)  # 200 means good!

# Step 2: Make "soup" to search easily
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Find all books (they are in <article class="product_pod">)
books = soup.find_all('article', class_='product_pod')

# Step 4: Loop through first 5 books and get title + price
print("\nFirst 5 Books:")
for book in books[:10]:
    title = book.h3.a['title']  # Get book title
    price = book.find('p', class_='price_color').text  # Get price
    print(f"Title: {title}")
    print(f"Price: {price}")
    print("---")
```

Convert this into a table using Pandas

```python
import pandas as pd

# Collect data
data = []
for book in books[:10]:  # First 10 books
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    data.append({"Title": title, "Price": price})

# Make table and show
df = pd.DataFrame(data)
print(df)
```
