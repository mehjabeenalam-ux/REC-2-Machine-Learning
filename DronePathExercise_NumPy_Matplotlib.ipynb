Drone Path Exercise with NumPy Arrays

Create a NumPy Array for the starting position (0, 0).

```python
import numpy as np

drone_position_zero = np.array([0,0])
print(drone_position_zero)

#One array contains two values (x = 0, y = 0)
```

Creating a Drone Path using a 2D Array: A Path containes many drone positions and Each row ina an Array represents a single drone position.

```python
path = np.array([
    [0, 0],
    [1, 1],
    [2, 1],
    [3, 2]
])

print(path)
```

Understanding the shape, dimesnion and type of the 2D Array created for the above Drone Path

```python
print(path.shape)
print(path.ndim)
print(np.dtype)
```

Drone Paths from above:

Start → (0,0)
Move  → (1,1)
Move  → (2,1)
Move  → (3,2)

Use Indexing to extract the x and y coordinates of the above 2D Array and Split the drone path into coordinates

```python
x = path[:, 0]
y = path[:, 1]

print(x)
print(f"X Coordinates of the Drone Path {x}")

print(y)
print(f"Y Coordinates of the Drone Path {y}")
```

Visualizing the Drone Path using Matplotlib Library

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel("X position")
plt.ylabel("Y position")
plt.title("Drone Flight Path")
plt.show()
```

Add a return path atthe origin (0,0)

```python
import numpy as np
return_path = np.array([[0, 0]]) #See the waythe array is written for the return path
print(return_path)
```

Total Path from Starting Point Till the Return Path using VStack

```python
path = np.vstack((path, return_path))
print(path)
```

Use Indexing to extract the x and y coordinates and split the entire drone path into coordinates

```python
x = path[:, 0]
y = path[:, 1]

print(x)
print(f"X Coordinates of the Entire Drone Path {x}")

print(y)
print(f"Y Coordinates of the Entire Drone Path {y}")
```

Visual Representation of the Entire Drone Path

```python
x = path[:, 0]
y = path[:, 1]

plt.plot(x, y)
plt.scatter(x, y)

plt.scatter(x[0], y[0])    # Start
plt.scatter(x[-1], y[-1])  # End (back to origin)

plt.xlabel("X position")
plt.ylabel("Y position")
plt.title("Complete Drone Flight Path (Return to Origin)")
plt.show()
```

Measuring Distance

Step 1: “Subtract each position from the next one.”

Each row = movement vector

```python
diff = np.diff(path, axis=0)
print(diff)
```

Step 2: Distance per step

= Calculate the length of each movement.

Code Explanation:-

- “Find how far the drone moved at each step.”

- diff → movement vectors

- np.linalg.norm → distance formula

- axis=1 → calculate distance row by row

```python
step_distances = np.linalg.norm(diff, axis=1)
print(step_distances)
for i, d in enumerate(step_distances, start=1):
    print(f"Step {i} distance: {d:.2f}")
```

Add each of these distances obtained to calculate the total distance

```python
Total_Distance = sum(step_distances)
print(Total_Distance)
print(round(Total_Distance, 2))
```

Adding Simple Animation

```python
import time

plt.figure()

for i in range(1, len(path) + 1):
    plt.clf()  # clear previous frame

    # plot path up to current step
    plt.plot(x[:i], y[:i], marker='o')

    # mark start and current position
    plt.scatter(x[0], y[0])      # start
    plt.scatter(x[i-1], y[i-1])  # current drone position

    plt.xlabel("X position")
    plt.ylabel("Y position")
    plt.title("Drone Path Animation")
    plt.grid(True)

    plt.pause(1.0)  # pause time (seconds)

plt.show()
```

Speed Management manipulate plt.pause(0.7) at 0.3, 0.7 and 1.0

Animating the entire drone with a drone image in the graph

```python
plt.figure(figsize=(6, 6))

def draw_drone(x_pos, y_pos):
    plt.imshow(
        img,
        extent=(
            x_pos - 0.3, x_pos + 0.3,
            y_pos - 0.3, y_pos + 0.3
        )
    )

for i in range(len(path)):
    plt.clf()  # clear previous frame

    # Draw path so far
    plt.plot(x[:i+1], y[:i+1], linestyle='--', marker='o')

    # Draw drone image at current position
    draw_drone(x[i], y[i])

    # Start & End markers
    plt.scatter(x[0], y[0])
    plt.scatter(x[-1], y[-1])

    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.title("Animated Drone Returning to Ground Zero!")
    plt.grid(True)
    plt.axis("equal")

    plt.pause(1.0)


plt.show()
```

Saving this drone path as an image file to reuse it in implementing Flask Project

```python
plt.figure(figsize=(6, 6))

# Draw full path
plt.plot(x, y, linestyle='--', marker='o')

# Draw drone at final position
draw_drone(x[-1], y[-1])

# Start & End
plt.scatter(x[0], y[0])
plt.scatter(x[-1], y[-1])

plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("Drone Flight Path (Mission Complete)")
plt.grid(True)
plt.axis("equal")

plt.savefig("drone_path.png")  # 🔥 IMPORTANT LINE
plt.show()
```
